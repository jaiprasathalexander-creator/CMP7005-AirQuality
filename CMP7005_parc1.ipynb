# Data Selection & Handling

**Station Justification (Xu & Zhang, 2004; Yao et al., 2015):**
- **Dongsi & Guanyuan** are located within the 5th Ring Road in central Beijing — classified as urban/inner stations with high traffic and dense residential activity.
- **Changping & Huairou** are located in the northern suburban belt, characterised by lower population density and proximity to semi-rural areas, classified as outer/suburban stations.

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# ── Load the four station datasets ──────────────────────────────────────
dongsi    = pd.read_csv('CMP7005-AirQuality/PRSA_Data_Dongsi_20130301-20170228.csv')
guanyuan  = pd.read_csv('CMP7005-AirQuality/PRSA_Data_Guanyuan_20130301-20170228.csv')
changping = pd.read_csv('CMP7005-AirQuality/PRSA_Data_Changping_20130301-20170228.csv')
huairou   = pd.read_csv('CMP7005-AirQuality/PRSA_Data_Huairou_20130301-20170228.csv')

# ── Add station identifier ───────────────────────────────────────────────
dongsi['station']    = 'Dongsi'
guanyuan['station']  = 'Guanyuan'
changping['station'] = 'Changping'
huairou['station']   = 'Huairou'

# ── Merge into a single unified dataset ─────────────────────────────────
df = pd.concat([dongsi, guanyuan, changping, huairou], ignore_index=True)

# ── Create datetime column from year/month/day/hour ─────────────────────
df['datetime'] = pd.to_datetime(df[['year', 'month', 'day', 'hour']])
df = df.sort_values('datetime').reset_index(drop=True)

print(f'Merged dataset shape: {df.shape}')
print(f'Date range: {df["datetime"].min()} → {df["datetime"].max()}')
df['station'].value_counts()
```

# Data Preprocessing

```python
# ── Remove duplicates ────────────────────────────────────────────────────
before = len(df)
df = df.drop_duplicates()
print(f'Duplicates removed: {before - len(df)}')

# ── Handle missing values with forward-fill then backward-fill ──────────
num_cols = ['PM2.5','PM10','SO2','NO2','CO','O3','TEMP','PRES','DEWP','RAIN','WSPM']
df[num_cols] = df[num_cols].ffill().bfill()
df['wd']     = df['wd'].ffill().bfill()

print(f'Remaining missing values: {df.isnull().sum().sum()}')

# ── Feature Engineering ─────────────────────────────────────────────────
df['year']  = df['datetime'].dt.year
df['month'] = df['datetime'].dt.month
df['day']   = df['datetime'].dt.day
df['hour']  = df['datetime'].dt.hour

df['season'] = df['month'].map({
    12:'Winter',1:'Winter',2:'Winter',
    3:'Spring', 4:'Spring', 5:'Spring',
    6:'Summer', 7:'Summer', 8:'Summer',
    9:'Autumn',10:'Autumn',11:'Autumn'
})

df['station_type'] = df['station'].map({
    'Dongsi':'Urban','Guanyuan':'Urban',
    'Changping':'Suburban','Huairou':'Suburban'
})

def aqi_category(pm25):
    if   pm25 <= 12:    return 'Good'
    elif pm25 <= 35.4:  return 'Moderate'
    elif pm25 <= 55.4:  return 'Unhealthy for Sensitive Groups'
    elif pm25 <= 150.4: return 'Unhealthy'
    elif pm25 <= 250.4: return 'Very Unhealthy'
    else:               return 'Hazardous'

df['AQI_Category'] = df['PM2.5'].apply(aqi_category)

print('\nFeature engineering complete.')
print(df[['datetime','season','station_type','AQI_Category']].head())
```
